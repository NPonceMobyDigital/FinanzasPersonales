# [dbo].[VistaPerfiles]

```sql
/****** Object:  View [dbo].[VistaPerfiles]    Script Date: 22/12/2022 8:27:30 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE VIEW [dbo].[VistaPerfiles]
AS
SELECT        dbo.Perfiles.ID_Saldo AS Saldo, dbo.Perfiles.ID_Ingreso AS Ingreso, dbo.Perfiles.ID_IngresoFuturo AS IngresoFuturo, dbo.Perfiles.ID_Egreso AS Egreso, dbo.Perfiles.ID_EgresoFuturo AS EgresoFuturo, 
                         dbo.Perfiles.ID_SaldoFuturo AS [Saldo Futuro], dbo.Perfiles.Nombre, dbo.Perfiles.Usuario
FROM            dbo.Egresos INNER JOIN
                         dbo.Perfiles ON dbo.Egresos.ID_Egreso = dbo.Perfiles.ID_Egreso INNER JOIN
                         dbo.EgresosFuturos ON dbo.Perfiles.ID_EgresoFuturo = dbo.EgresosFuturos.Id_EgresosFuturos INNER JOIN
                         dbo.Ingresos ON dbo.Perfiles.ID_Ingreso = dbo.Ingresos.ID_Ingreso INNER JOIN
                         dbo.IngresosFuturos ON dbo.Perfiles.ID_IngresoFuturo = dbo.IngresosFuturos.ID_IngresosFuturos INNER JOIN
                         dbo.Saldos ON dbo.Perfiles.ID_Saldo = dbo.Saldos.ID_Saldo INNER JOIN
                         dbo.SaldosFuturos ON dbo.Perfiles.ID_SaldoFuturo = dbo.SaldosFuturos.ID_MontoFuturo
GO
```
